## seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

np.random.seed(0)
x = np.random.randn(100)
```

```python
sns.rugplot(x)
```

```python
sns.kdeplot(x)  # 누적분포를 선으로 표현
```

```python
sns.distplot(x, kde=True, rug=True)  # 히스토그램과 비슷
```

```python
tips = sns.load_dataset('tips')  # eaborn이 제공하는 데이터셋 불러오기
tips.head()
```

```python
tips = sns.load_dataset('tips')
sns.jointplot(x='total_bill', y='tip', data=tips)  # scatter plot과 비슷
```

```python
iris = sns.load_dataset('iris')
sns.jointplot('sepal_width','petal_length', data=iris, kind='kde', \
             space=0, color='g')
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)  # regression plot (회귀분석) + 스캐터 플롯 + smoker 한번에
```

## residplot

```python
sns.residplot(x='total_bill', y='tip', data=tips)  # 잔차항(회귀분석의 에러)을 스캐터 플롯으로 그림
```

## lmplot

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker',data=tips)  
# 스캐터 + 회귀 직선 + hue값 설정(smoker=yes와 no 각각 나눠서)
```

```python
sns.lmplot(x='total_bill', y='tip', col='smoker', data=tips)
```

## barplot

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips)  # bar plot
```

## countplot

```python
titanic = sns.load_dataset('titanic')
sns.countplot(x='class', hue='who', data=titanic)  # 개수 표현 (hue 설정하여 상세하게 카운트 가능)
```

## boxplot, pointplot

```python
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips)  # 박스 플롯
```

```python
sns.pointplot(x='time',y='total_bill',hue='smoker', data=tips, dodge=True)  # 증감 표현
```

## violinplot, stripplot

```python
# bar plot의 다른 형태
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, palette='muted')
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', 
              data=tips, palette='Set2', split=True,
              scale='count', inner='quartile')
```

```python
sns.stripplot(x='day', y='total_bill', hue='smoker', 
              data=tips, jitter=True,
             palette='Set2', dodge=True)
```

```python
sns.boxplot(x='tip', y='day', data=tips, whis=np.inf)
sns.stripplot(x='tip', y='day', data=tips, jitter=True, color='.3')
```

```python
sns.swarmplot(x='day', y='total_bill', hue='sex', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, inner=None)
sns.swarmplot(x='day', y='total_bill', data=tips, color='White', edgecolor='gray')
```

## heatmap

```python
flights = sns.load_dataset('flights')
flights = flights.pivot('month', 'year', 'passengers')
sns.heatmap(flights, annot=True, fmt='d')
```
